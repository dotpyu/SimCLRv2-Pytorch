```python
from collections import OrderedDict

simclr_new_dict = OrderedDict()


translate = {'net.0':'layer1.0', 'net.1':'layer1.1', 'net.2':'layer2.0', 'net.3':'layer3.0', 'net.4':'layer4.0'}

for name, val in simclr.items():
    if name[:4] == 'net.':
        simclr_new_dict[translate[name[:5]]+name[5:]] = val
    else:
        simclr_new_dict[name] = val
```

```python
torch.save(simclr_new_dict, 'adjusted_simclr_r50.pth')
```
